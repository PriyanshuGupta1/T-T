```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
df
```

```python
cols = ['Age', 'EstimatedSalary', 'Purchased']
```

```python
for col in cols:
    le = preprocessing.LabelEncoder()
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
df.Purchased.value_counts()
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns = ['Purchased']), df['Purchased'], test_size = 0.3)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
algos = []
accuracy = []
recall = []
specificity=[]
precision = []
f1Score = []
```

```python
algo = "Logistic Regression"
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(algo)

print(confusion_matrix(y_test, y_pred), '\n')
cm1=confusion_matrix(y_test, y_pred)

acc = accuracy_score(y_test, y_pred) 
print('Accuracy:', acc)

rec = recall_score(y_test, y_pred) 
print('Recall:', rec)

sp1=cm1[0][0]/(cm1[0][0]+cm1[0][1])
print('Specificity:', sp1)

pre = precision_score(y_test, y_pred) 
print('Precision:', pre)

f1s = f1_score(y_test, y_pred) 
print('F score:', f1s)

algos.append(algo)
accuracy.append(acc)
specificity.append(sp1)
recall.append(rec)
precision.append(pre)
f1Score.append(f1s)
print("\n")
```

```python
algo="SVM"
svm=SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
print(algo)

print(confusion_matrix(y_test, y_pred), '\n')
cm4=confusion_matrix(y_test, y_pred)

acc = accuracy_score(y_test, y_pred) 
print('Accuracy:', acc)

rec = recall_score(y_test, y_pred) 
print('Recall:', rec)

sp4=cm4[0][0]/(cm4[0][0]+cm4[0][1])
print('Specificity:', sp4)

pre = precision_score(y_test, y_pred) 
print('Precision:', pre)

f1s = f1_score(y_test, y_pred) 
print('F1 score:', f1s)

algos.append(algo)
accuracy.append(acc)
specificity.append(sp4)
recall.append(rec)
precision.append(pre)
f1Score.append(f1s)
print("\n")
```

```python
algo = "Decision Tree"
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(algo)

print(confusion_matrix(y_test, y_pred), '\n')
cm3=confusion_matrix(y_test, y_pred)

acc = accuracy_score(y_test, y_pred)
print('Accuracy:', acc)

rec = recall_score(y_test, y_pred)
print('Recall:', rec)

sp3=cm3[0][0]/(cm3[0][0]+cm3[0][1])
print('Specificity:', sp1)

pre = precision_score(y_test, y_pred) 
print('Precision:', pre)

f1s = f1_score(y_test, y_pred) 
print('F score:', f1s)

algos.append(algo)
accuracy.append(acc)
specificity.append(sp3)
recall.append(rec)
precision.append(pre)
f1Score.append(f1s)
print("\n")
```
